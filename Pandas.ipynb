```python
# CHECKING THE DATATYPE
type(df)
```

```python
# KNOWING ABOUT COLUMN
df.columns
```

```python
# CONVERTING DATAFRAME/SERIES INTO ARRAY
df.to_numpy()
# 'to_numpy()' method is used to convert Pandas Dataframe or Series into NumPy array.
```

```python
# CREATING CSV FILE FROM DATAFRAME
df.to_csv('code.csv')
# csv file will create a spreadsheet file to show data
```

```python
# STARTING RECORDS
df.head()
```

```python
# DROP() FUNCTION
# when we have to drop or delete a row or a column we use (axis) to define them
# for deleting a column, axis = 1 is used to define column
# for deleting a row, axis = 0 is used to define row
```

```python
# DELETING A COLUMN
# df.drop(column_name,axis = 1)
df = df.drop('a',axis = 1)
```

```python
df
```

```python
# DELETING A ROW
# df.drop(row_name,axis = 0)
df = df.drop(0,axis = 0)
```

```python
df
```

```python
# ILOC [works on the index]
# ILOC is used to select and access data in dataframme or series using integer based indexing
```

```python
# [row_num , column_num]
df.iloc[0,1]
```
